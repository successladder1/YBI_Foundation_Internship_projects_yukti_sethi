Features in the dataset:

1. GRE Scores (290 to 340)
2. TOEFL Scores (92 to 120)
3. University Rating (1 to 5)
4. Statement of Purpose (1 to 5)
5. Letter of Recommendation Strength (1 to 5)
6. Undergraduate CGPA (6.8 to 9.92)
7. Research Experience (0 or 1)
8. Chance of Admit (0.34 to 0.97)

```python
# import library
import pandas as pd
import numpy as array
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# import data
admission = pd.read_csv('https://github.com/ybifoundation/Dataset/raw/main/Admission%20Chance.csv')
```

```python
# view data
admission.head()
```

```python
# info of data
admission.info()
```

```python
# check for missing value
admission.isna().sum()
```

```python
# check for categories
admission.nunique()
```

```python
# visualize pairplot
sns.pairplot(admission)
```

```python
# columns name
admission.columns
```

```python
# define y
y=admission['Chance of Admit '] #target
```

```python
# define X
x=admission[['GRE Score', 'TOEFL Score', 'University Rating', ' SOP', 
       'LOR ', 'CGPA', 'Research']] #features
```

```python
# split data
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=.30, random_state=2529)
```

```python
# verify shape
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
# select model
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
# train model
model.fit(x_train,y_train)
```

```python
# predict with model
c=model.intercept_
print(c) #intercept
m=model.coef_
print(m) #slope
y_pred=model.predict(x_test)
print(y_pred) #predicted value of target using ML algo
```

```python
# model evaluation
from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error, mean_squared_error
```

```python
# model MAE
mean_absolute_error(y_test,y_pred)
```

```python
# model MAPE
mean_absolute_percentage_error(y_test, y_pred)
```

```python
# model MSE
mean_squared_error(y_test, y_pred)
```
