There are 14 attributes in each case of the dataset. They are:
1. CRIM - per capita crime rate by town
2. ZN - proportion of residential land zoned for lots over 25,000 sq.ft.
3. INDUS - proportion of non-retail business acres per town.
4. CHAS - Charles River dummy variable (1 if tract bounds river; 0 otherwise)
5. NOX - nitric oxides concentration (parts per 10 million)
6. RM - average number of rooms per dwelling
7. AGE - proportion of owner-occupied units built prior to 1940
8. DIS - weighted distances to five Boston employment centres
9. RAD - index of accessibility to radial highways
10. TAX - full-value property-tax rate per 10,000 dolar
11. PTRATIO - pupil-teacher ratio by town
12. B - 1000(Bk - 0.63)^2 where Bk is the proportion of blacks by town
13. LSTAT - % lower status of the population
14. MEDV - Median value of owner-occupied homes in 1000 dollars

```python
# import library
import pandas as pd
import numpy as array
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# import data
house = pd.read_csv('https://github.com/ybifoundation/Dataset/raw/main/Boston.csv')
```

```python
# view data
house.head()
```

```python
# info of data
house.info()
```

```python
# summary statistics
house.describe()
```

```python
# check for missing value
house.isna().sum()
```

```python
# check for categories
house.nunique()
```

```python
# visualize pairplot
sns.pairplot(house)
```

```python
# columns name
house.columns
```

```python
# define y
y=house['MEDV']
```

```python
# define X
X=house[['CRIM', 'ZN', 'INDUS', 'CHAS', 'NX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX',
       'PTRATIO', 'B', 'LSTAT']]
```

```python
# split data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(X,y,test_size=.30, random_state=2529)
```

```python
# verify shape
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
# select model
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
# train model
model.fit(X_train,y_train)
```

```python
# predict with model
y_pred=model.predict(X_test)
```

```python
# model evaluation
from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error, mean_squared_error
```

```python
# model MAE
mean_absolute_error(y_test,y_pred)
```

```python
# model MAPE
mean_absolute_percentage_error(y_test,y_pred)
```

```python
# model MSE
mean_squared_error(y_test,y_pred)
```
